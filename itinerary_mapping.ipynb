```python
#get list of unique course number
#input [df]: sort_values('Course number')
#output [array]: course_nr

course_nr = freq.sort_values('Course number')
course_nr = course_nr['Course number'].unique()
course_nr
```

#**Step 1: Create Base DF**

```python
#create df with columns: "course_nr", "station", "previous_stat", "next_stat", "IN", "OUT"
#input : course_nr, freq
#output: itinerary

it_df = pd.DataFrame(columns =["course_nr", "station", "previous_stat", "next_stat"])

for i in course_nr:
  for m, n in freq.iterrows():
    if i == n[1]:
      empty = pd.DataFrame(columns =["course_nr", "station", "previous_stat", "next_stat"])
      empty.append([""])
      empty.at[0, ["course_nr", "station"]] = n[1], n[0]
      empty.at[0, [n[2]]] = n[3]
      it_df = it_df.append(empty, ignore_index = True)

#group df by course number and station, in order t merge IN and OUT into one row
from numpy import NaN
itinerary_df = it_df.groupby(['course_nr', 'station']).sum()
itinerary = itinerary_df.drop(columns = NaN, axis=1)

itinerary
```

```python
#clean the cloumn; correct the bad data ('IN','OUT'; numbers are e0 and e3)

itinerary = itinerary_df.drop(columns = NaN, axis=1)

itinerary['previous_stat'] = ""
itinerary['next_stat'] = ""

itinerary['IN'] = itinerary['IN'].str.replace('-','0')
itinerary['OUT'] = itinerary['OUT'].str.replace('-','0')

itinerary['IN'] = [float(x['IN']) * 1000 if float(x['IN']) % 1 != 0 else x['IN'] for i, x in itinerary.iterrows()] # * 1000 if residuals
itinerary['OUT'] = [float(x['OUT']) * 1000 if float(x['OUT']) % 1 != 0 else x['OUT'] for i, x in itinerary.iterrows()]


itinerary
```

#**Step 2: Reduce Dimensionality**

```python
#reduce dimensionality of fahrplan for convenient looping
fahrplan = pd.melt(fahr_p, id_vars=['Kursnummer'], var_name = 'course_nr', value_name = 'halt', value_vars=['401', '7',	'407',	'11',	'13',	'15',	'415',	'21',	'23',	'25',	'425',	'27',	'427',	'29',	'33', '1', '403'], ignore_index=False)
fahrplan.rename(columns={'Kursnummer':'station'}, inplace=True)
fahrplan
```

#**Step 3: Map "previous station" and "next station" in Base DF**

```python
#map station with previous station and next station

"""
itinerary:
i = index
r = course_nr, station, previous_stat, next_stat, total, IN, OUT

fahrplan:
n = index
m = station, course_nr, halt
"""

for i in course_nr:
  if fahrplan[(fahrplan['course_nr'] == i) & (fahrplan['halt'] == 'x')].empty == False:
    subset = (fahrplan[(fahrplan['course_nr'] == i) & (fahrplan['halt'] == 'x')])
    subset.reset_index(inplace = True)
    length = len(subset)
    print(subset)
    print(f'\nlength: {length}')
    for n, r in subset.iterrows():
      print(f'index: {n}\n')
      print(f'Station: {r["station"]}\n')

      if n == 0:
        print('Is First Station')
        itinerary.at[(str(r['course_nr']), r['station']),'previous_stat'] = "start"  #access cell value df.at[row, column]
      else:
        itinerary.at[(str(r['course_nr']), r['station']),'previous_stat'] = subset['station'][(n-1)]
        print("Previous Station: ", subset['station'][(n-1)])

      if n+1 == length:
        print('Is Last Station')
        itinerary.at[(str(r['course_nr']), r['station']),'next_stat'] = "end"
      else:
        print(f'Next Station: {subset["station"][(n+1)]}')
        itinerary.at[(r['course_nr'], r['station']),'next_stat'] = subset['station'][(n+1)]

      print('END\n')
```

#**Step 4: Map coordinates of station and next station**

```python
#load station coordinates into pd df

coord = pd.read_excel('/content/drive/My Drive/Colab Notebooks/geo_data/koordinaten.xlsx')
coord
```

```python
itinerary[itinerary['previous_stat'] != ""]
```

```python
# add column to main df: 'station_long', 'station_lat', 'next_long', 'next_lat'
#get rid of column missing data 
itinerary = itinerary[(itinerary['previous_stat'] != "") & (itinerary['IN'].astype(float) != 0.0)]

itinerary['station_long'] = ""
itinerary['station_lat'] = ""
itinerary['next_long'] = ""
itinerary['next_lat'] = ""

for i, r in itinerary.iterrows():
  itinerary['station_long'][i] = str(coord['Long'][coord['Station'] == i[1]].values).strip('[]')
  itinerary['station_lat'][i] = str(coord['Lat'][coord['Station'] == i[1]].values).strip('[]')
  itinerary['next_long'][i] = str(coord['Long'][coord['Station'] == r['next_stat']].values).strip('[]')
  itinerary['next_lat'][i] = str(coord['Lat'][coord['Station'] == r['next_stat']].values).strip('[]')

itinerary
```

```python
#reset index and ad column 'load'

itinerary_re_i = itinerary.reset_index(level=[0,1])
itinerary_re_i["load"] = 0
itinerary_re_i
```

```python
#calculate passenger load for all starting stations; load = IN

for i, r in itinerary_re_i[itinerary_re_i['previous_stat'] == "start"].iterrows():
 itinerary_re_i['load'][i] = int(itinerary_re_i['IN'][i])

itinerary_re_i
```

```python
#calculate passenger load for all NON starting stations; load = 'previous load' + (IN - OUT)
#because df order is not correct: innerloop breaks if 0 in prev_load, outerloop stops if no 0 in load

while 0 in itinerary_re_i['load'].values:
  print(0 in itinerary_re_i['load'])
  print('_' * 60)
  for i, r in itinerary_re_i[itinerary_re_i['previous_stat'] != "start"].iterrows():
      prev_load = itinerary_re_i[(itinerary_re_i['course_nr'] == r['course_nr']) & (itinerary_re_i['station'] == r['previous_stat'])]['load'] #get load value of previous station
      if prev_load.values != 0:
        itinerary_re_i['load'][i] = (int(prev_load.values) + (int(r['IN']) - int(r['OUT'])))
        print(f'calculated load {itinerary_re_i["load"][i]}')

```
